```python
ds.fillna(ds.median(), inplace=True)
print(ds.isna().sum())
```

```python
#8.	Каковы медиана и стандартное отклонение платежей (Fare)? Округлите до 
#2 десятичных знаков
print(round(ds["Fare"].median(),2))
print(round(ds["Fare"].std(),2))
```

```python
#9.	Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет? 
#Каковы доли выживших в обеих группах?
ds_surv=ds[ds['Survived']==1]['Survived'].sum()
#print(ds_surv)
ds30_1=ds[ds['Age']<30]['Survived'].sum()
#print(ds30_1)
ds60_1=ds[ds['Age']>60]['Survived'].sum()
#print(ds60_1)
print(ds30_1>ds60_1)
print(round(ds30_1/ds_surv,2))
print(round(ds60_1/ds_surv,2))
```

```python
#10.Правда ли, что женщины выживали чаще мужчин? Каковы доли выживших в 
#обеих группах?
```

```python
ds_male=ds[ds['Sex']=='male']['Survived'].sum()
ds_female=ds[ds['Sex']=='female']['Survived'].sum()
print(ds_female>ds_male)
print(round(ds_female/ds_surv, 2))
print(round(ds_male/ds_surv, 2))
```

```python
#13.Как отличается средний возраст мужчин / женщин в зависимости от класса 
#обслуживания? Выберите верные утверждения:
#В среднем мужчины 1-го класса старше 40 лет
print(ds[(ds['Pclass']==1) & (ds['Sex']=='male')]['Age'].mean() > 40)
```

```python
#В среднем женщины 1-го класса старше 40 лет
print(ds[(ds['Pclass']==1) & (ds['Sex']=='female')]['Age'].mean() > 40)
```

```python
#Мужчины всех классов в среднем старше женщин того же класса
male1=ds[(ds['Pclass']==1) & (ds['Sex']=='male')]['Age'].mean()
female1=ds[(ds['Pclass']==1) & (ds['Sex']=='female')]['Age'].mean()
male2=ds[(ds['Pclass']==2) & (ds['Sex']=='male')]['Age'].mean()
female2=ds[(ds['Pclass']==2) & (ds['Sex']=='female')]['Age'].mean()
male3=ds[(ds['Pclass']==3) & (ds['Sex']=='male')]['Age'].mean()
female3=ds[(ds['Pclass']==3) & (ds['Sex']=='female')]['Age'].mean()
print((male1>female1) and (male2>female2) and (male3>female3))
```

```python
#В среднем люди в 1 классе старше, чем во 2-ом, а те старше представителей 3-го класса
print(ds[(ds['Pclass']==1)]['Age'].mean() >
      ds[(ds['Pclass']==2)]['Age'].mean() >
      ds[(ds['Pclass']==3)]['Age'].mean() )
```
